```python
import yfinance as yf #uses yahoo finance API to get daily stock and index prices
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
#All zeros in both columns
del sp500['Dividends']
del sp500['Stock Splits']
```

```python
#Target variable = stock value going up or down

#Price of tomorrow is price of close shifted one day
sp500['Tomorrow'] = sp500['Close'].shift(-1)
```

```python
sp500
```

```python
sp500['Target'] = (sp500['Tomorrow'] > sp500['Close']).astype(int)
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy() #Use more recent stock information
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100, min_samples_split = 100, random_state = 1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ['Close', 'Volume', 'Open', 'High', 'Low']
model.fit(train[predictors], train['Target'])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd 

preds = pd.Series(preds, index=test.index)
```

```python
preds
```

```python
precision_score(test['Target'], preds)
```

```python
combined = pd.concat([test['Target'], preds], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis=1)
    return combined
```

```python
#implement backtesting
#start of 2500 (250 days/year in dataset over 10 years)
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
predictions['Target'].value_counts()/predictions.shape[0]
```

```python
intervals = [2, 5, 60, 250, 1000] #Calculate rolling averages across these day intervals
new_predictors = []

for interval in intervals:
    rolling_averages = sp500.rolling(interval).mean()
    ratio_column = f"Close_Ratio_{interval}"
    trend_column = f"Trend_{interval}"
    sp500[ratio_column] = sp500['Close']/rolling_averages['Close']
    sp500[trend_column] = sp500.shift(1).rolling(interval).sum()['Target'] #rolling sum of target
    new_predictors+=[ratio_column, trend_column]
```

```python
sp500
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state = 1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict_proba(test[predictors])[:,1] #probability stock price will go up or down
    preds[preds >= .6] = 1 #model should be over 60% confident stock price will go up
    preds[preds <= .6] = 0
    preds = pd.Series(preds, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
#Stock went up 53% of days, stock goes up on 57% of days that the model says to buy
#Model has predictive power --> success!
```
